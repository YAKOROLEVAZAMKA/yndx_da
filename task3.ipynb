```python
def calculate_surpass(df, col, col_norm, col_final):
    df[col_final] = 0
    
    df[col_final] = df[col] - df[col_norm]
    df[col_final] = df[col_final].apply(lambda x: 0 if x < 0 else x)
```

–ê–±–æ–Ω–µ–Ω—Ç–∞–º —Ç–∞—Ä–∏—Ñ–∞ <b>'smart'</b> —Å—Ç–æ–∏—Ç –∑–∞–¥—É–º–∞—Ç—å—Å—è –≤—Å—ë –ª–∏ –æ–Ω–∏ –¥–µ–ª–∞—é—Ç –ø—Ä–∞–≤–∏–ª—å–Ω–æ.

```python
data_analyze = data.groupby(['month', 
                             'tariff', 
                             'rub_per_message', 
                             'rub_per_minute', 
                             'rub_per_gb']).agg({'user_id': 'count',
                                            'total_payed': 'sum',
                                            'overpay': 'sum',
                                            'minutes_overuse': 'sum',
                                            'sms_overuse': 'sum',
                                            'mb_overuse': 'sum'
                                           }).reset_index().rename(columns={'user_id':'total_users'})

data_analyze['overpay_minutes'] = data_analyze['minutes_overuse'] * data_analyze['rub_per_minute']
data_analyze['overpay_sms'] = data_analyze['sms_overuse'] * data_analyze['rub_per_message']
data_analyze['overpay_mb'] = data_analyze['mb_overuse'] * (data_analyze['rub_per_gb'] / 1024)

data_analyze['month'] = data_analyze['month'].astype(int)


data_analyze['overpay_percent'] = round(data_analyze['overpay'] * 100 / data_analyze['total_payed'], 2)
data_analyze['overpay_minutes_percent'] = round(data_analyze['overpay_minutes'] * 100 / data_analyze['overpay'], 2)
data_analyze['overpay_sms_percent'] = round(data_analyze['overpay_sms'] * 100 / data_analyze['overpay'], 2)
data_analyze['overpay_mb_percent'] = round(data_analyze['overpay_mb'] * 100 / data_analyze['overpay'], 2)

data_analyze.fillna(0, inplace=True)

data_analyze = data_analyze[['month', 
                             'tariff', 
                             'total_users',
                             'total_payed',
                             'overpay',
                             'overpay_percent',
                             'overpay_minutes_percent', 
                             'overpay_sms_percent',
                             'overpay_mb_percent']]
                            
```

```python
data_analyze
# –≠–¢–û –¢–ê–ë–õ–ò–¶–ê –ê–ù–ê–õ–ò–ó–ê –ü–ï–†–ï–ü–õ–ê–¢ –í –ü–†–û–¶–ï–ù–¢–ù–û–ú –°–û–û–¢–ù–û–®–ï–ù–ò–ò!
```

```python
def median_payments(df, tarifflist, cols):
    for tariff in tarifflist:
        for col in cols:
            print(f"–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–ø–ª–∞—Ç –ø–æ —Ç–∞—Ä–∏—Ñ—É {tariff} = {round(df[df['tariff'] == tariff][col].median(), 2)}% –∑–∞ {cols[col]}")
        print('**********')
        
median_payments(data_analyze, 
                ['smart', 'ultra'], 
                {'overpay_minutes_percent': '–º–∏–Ω—É—Ç—ã', 
                 'overpay_sms_percent': 'sms',
                 'overpay_mb_percent': '–º–µ–≥–∞–±–∞–π—Ç—ã'})
```

### 4. –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑

- —Å—Ä–µ–¥–Ω—è—è –≤—ã—Ä—É—á–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ç–∞—Ä–∏—Ñ–æ–≤ ¬´–£–ª—å—Ç—Ä–∞¬ª –∏ ¬´–°–º–∞—Ä—Ç¬ª —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è
- —Å—Ä–µ–¥–Ω—è—è –≤—ã—Ä—É—á–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ –ú–æ—Å–∫–≤—ã –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –≤—ã—Ä—É—á–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ –¥—Ä—É–≥–∏—Ö —Ä–µ–≥–∏–æ–Ω–æ–≤

–ü–æ—è—Å–Ω–∏—Ç–µ:

- –∫–∞–∫ –≤—ã —Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–ª–∏ –Ω—É–ª–µ–≤—É—é –∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É—é –≥–∏–ø–æ—Ç–µ–∑—ã
- –∫–∞–∫–æ–π –∫—Ä–∏—Ç–µ—Ä–∏–π –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑ –∏ –ø–æ—á–µ–º—É




```python
ultra = data[data['tariff'] == 'ultra']
smart = data[data['tariff'] == 'smart']
```

```python
def drop_anomalies(df, col):
    q1 = df[col].quantile(0.25)                 
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    iqr_test = (df[col] >= (q1 - 1.5 * iqr)) & (df[col] <= (q3 + 1.5 * iqr))

    return df.loc[iqr_test][col]
```

```python
ultra = drop_anomalies(ultra, 'total_payed')

smart = drop_anomalies(smart, 'total_payed')
```
