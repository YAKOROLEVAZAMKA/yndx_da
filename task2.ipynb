```python
# –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –æ–±—ä–µ–∫—Ç –≤ –¥–∞—Ç—É
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format="%Y-%m-%d")

# –≥–æ–¥ - –º–µ—Å—è—Ü - –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏

data['year'] = data['first_day_exposition'].dt.year
data['month'] = data['first_day_exposition'].dt.month
data['weekday'] = data['first_day_exposition'].dt.dayofweek
```

```python
# –∏–∑–º–µ–Ω–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö + –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ
def convert_to_int(df, arr):
    for col in arr:
        df[col] = df[col].fillna(0)
        df[col] = df[col].astype(int)
        
cols_to_int = ['days_exposition', 
               'floors_total', 
               'balcony_amount', 
               'parks_amount_within_3km', 
               'ponds_amount_within_3km']

convert_to_int(data, cols_to_int)

def change_distance(df, arr, dist=1000):
    for col in arr:
        df[col] = df[col].fillna(0)
        df[col] = round(df[col] / dist, 2)
        df.rename(columns={col: col.replace('_m', '_km')}, inplace=True)

cols_to_km = ['airport_distance_m', 
              'city_center_distance_m', 
              'nearest_park_distance_m',
              'nearest_pond_distance_m']        
        
change_distance(data, cols_to_km)
```

```python
def table_metrics(df, df_clean, metric_arr, func='median'):
    df_final = None
    
    for metric in metric_arr:
        if df_final is None:
            df_final = pd.DataFrame(zip([metric], 
                                        [df[metric].agg(func=func)], 
                                        [df_clean[metric].agg(func=func)]),
                                   columns=['–ú–µ—Ç—Ä–∏–∫–∞', '–° –≤—ã–±—Ä–æ—Å–∞–º–∏', '–ë–µ–∑ –≤—ã–±—Ä–æ—Å–æ–≤']
                                   )
        else:
            df_temp = pd.DataFrame(zip([metric], 
                                       [df[metric].agg(func=func)], 
                                       [df_clean[metric].agg(func=func)]),
                                   columns=['–ú–µ—Ç—Ä–∏–∫–∞', '–° –≤—ã–±—Ä–æ—Å–∞–º–∏', '–ë–µ–∑ –≤—ã–±—Ä–æ—Å–æ–≤']
                                   ).reset_index(drop=True)
            df_final = pd.concat([df_final, df_temp])
    
    return df_final
```

```python
data_top10 = (data_clean['city_name'].value_counts() / data.shape[0] * 100).head(10)\
                                     .reset_index().rename(columns={'index': 'city_name',
                                                                    'city_name': 'quantity'})

data_top10_price = data_clean.query('city_name in @data_top10.city_name.unique()')\
                                                          .groupby(['city_name'])[['meter_price']]\
                                                          .median()\
                                                          .reset_index()\
                                                          .rename(columns={'meter_price': 'avg_meter_price'})\
                                                          .sort_values(by='avg_meter_price', ascending=False)\
                                                          .reset_index().rename(columns={'index': 'place'})
```

```python
data_merged = data_top10.merge(data_top10_price, left_on=['city_name'], right_on=['city_name'])
data_merged['place'] = abs(data_merged['place'] - 10)
data_merged = data_merged[['city_name', 'quantity', 'avg_meter_price', 'place']]
```
